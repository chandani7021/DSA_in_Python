```python
#Linear search

# is based on Brute force
#time complexity is O(n)
# No sorting required

def linear_search(arr, item):

    #apply loop to check if item is in array or not
    for i in range(len(arr)):

        #if array index i has item then return i(index value) or return -1
        if arr[i] == item:
            return i
        
    return -1
```

```python
linear_search(arr,167)
```

```python
#Binary search

#sorted array
# time complexity is O(log n)

def binary_search(arr, low, high, item):


    #print("low = ", low, "high= ", high, end = " ")
    
    if low <= high:
        #search
        mid = (low + high) // 2

      #  print("mid value : ", arr[mid])

        if arr[mid] == item:
            return mid
        elif arr[mid] > item:
            return binary_search(arr, low, mid-1, item)
        else:
            return binary_search(arr, mid+1 , high, item)
    else:
        return "Not found"
```

```python
print(binary_search(arr,0,len(arr) - 1,55))
```

```python
# sorting
# Q. How to detect whether the given array is sorted or not.

def is_sorted(arr):

    sorted = True

    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            sorted =  False
    return sorted
```

```python
# Bubble sort
# Time complexity = O(n ^ 2)

def bubble_sort(arr):

    #will apply loop for starting passes. 
    # passes is always 1 less then the length of array (eg. if len of array is 5 then we have to do 4 passes)

    for i in range(len(arr) - 1):
        # in each pass you have to compare the adjacent itmes 
        # so the no. of comparison will be ( if 5 ele. then 4 times comparison)
        # after 1 pass we will leave the last element because it is the largest and it is in its final position
        #to leave the last element we will subtract i
        for j in range(len(arr) - 1 - i):
            #Compare adjacent items
            if arr[j] > arr[j + 1]:
                # if j is bigger then j+1 then swap the items
                arr[j], arr[j+1] = arr[j + 1], arr[j]

    #to print uncomment it
    #print(arr)
    #it is to check the time taken by sort
    return arr
```

```python
# how to make Bubble sort adaptive
# in best case Time complexity - O(n) because one pass takes place


def adaptive_bubble_sort(arr):

    #will apply loop for starting passes. 
    # passes is always 1 less then the length of array (eg. if len of array is 5 then we have to do 4 passes)

    for i in range(len(arr) - 1):

        #introduce flag as 0
        flag = 0
        # in each pass you have to compare the adjacent itmes 
        # so the no. of comparison will be ( if 5 ele. then 4 times comparison)
        # after 1 pass we will leave the last element because it is the largest and it is in its final position
        #to leave the last element we will subtract i
        for j in range(len(arr) - 1 - i):
            #Compare adjacent items
            if arr[j] > arr[j + 1]:
                # if j is bigger then j+1 then swap the items
                arr[j], arr[j+1] = arr[j + 1], arr[j]

                # swap took place then change flag value to 1
                flag = 1

        #check if flag value is 0 or not
        if flag == 0:
            #if yes then break
            break

    print(arr)
```

```python
# selection sort
# Time complexity = O(n ^ 2)

#to understand the process nicely uncomment print

def selection_sort(arr):

    #will apply loop for starting passes. 
    # passes is always 1 less then the length of array (eg. if len of array is 5 then we have to do 4 passes)

    for i in range(len(arr) - 1):

        #print( i+1, "pass", end= " ")

        # assign the first ele as min and i
        min = i

       # print("current min is", arr[min])

        #leaving the ith element assign next element in j  to check if j is less then min
        for j in range(i+1, len(arr)):

           # print("Current item under observation ", arr[j])

            #check if j is less then min
            if arr[j] < arr[min]:

          #      print("Current item is less then min")
                #if yes then assign j as min
                min = j
         #       print("Now min has become ", arr[min])

        # swap the i and min   
        arr[i], arr[min] = arr[min], arr[i]
        #print(arr)
        #print("*"*50)

    #to print uncomment it
    #print(arr)
    #it is to check the time taken by sort
    return arr
```

```python
# Merge sort
# in this you have to give 2 sorted array and it will sort and merge them.
def merge_sorted(arr1, arr2):

    #intrduce 2 variables i and j for arr1 and arr2
    i = j = 0 

    #create array which is empty which will be the merged array 
    merged = []

    #apply loop while i and j is less then arr1 and 2
    while i < len(arr1) and j< len(arr2):

        #apply loop if i of array is less then j of array
        if arr1[i] < arr2[j]:
            #if yes then append i ele. in merged
            merged.append(arr1[i])
            #increment i
            i += 1
        else:
            #if no then appen j ele. in merged
            merged.append(arr2[j])
            #increment j
            j += 1

    # for the elements that are still left after the above loop,
    #to append them in merged array do while for i and j 
    # if the are less then arr1 and arr2 respectively
    #if yes then append them in merged
    while i < len(arr1):
        merged.append(arr1[i])
        i += 1

    while j < len(arr1):
        merged.append(arr2[j])
        j += 1

    # after loop return the merged array
    return merged
```

```python
arr1 = [9,18,27,36,75]
arr2 = [5,8,19,24,78]
```

```python
merge_sorted(arr1, arr2)
```

```python
#Now making a merge_sort function to sort single array
def merge_sort(arr):

    #Base case is if length of array is one then array is already sorted.
    if len(arr) == 1:
        return arr
    
    #if array length is not one then
    #divide the array as even as possible by calculating mod point
    mid = len(arr)//2

    #then make to arrays for elements that are on left of mid and on right of mid
    left = arr[:mid]
    right = arr[mid:]

    # now we have to sort both left and right array also
    # so we will do recursion( use merge_sort function) to sort arrays
    left = merge_sort(left)
    right = merge_sort(right)

    #now give this left and right sorted array to merge_Sorted function to merge the array and return the final merged array
    return merge_sorted(left, right)

```
